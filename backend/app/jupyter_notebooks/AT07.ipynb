```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = 500
pd.options.display.max_rows = 10000
```

```python
names = [
    'NumeroCredito', 
    'CodigoBien', 
    'FechaLiquidacion', 
    'CodigoContable', 
    'ClaseBien', 
    'TipoCliente', 
    'IdentificacionCliente', 
    'NombreRazonSocial', 
    'SituacionGarante', 
    'MontoInicial', 
    'MontoActual', 
    'MontoAvaluo', 
    'ValorMercado', 
    'FechaUltimoAvaluo', 
    'CodigoPeritoAvaluador',
]

converters = {
    'BranchId':str, 
}

parse_dates = [
    'FechaLiquidacion',
    'FechaUltimoAvaluo',
]
```

```python
AT07 = pd.read_csv('../vol/web/media/insumos/automatic/AT07_200108.txt',
                   sep='~', 
                   low_memory=False, 
                   encoding = "latin",
                   parse_dates=parse_dates, 
                   header=None,
                   names=names)
```

```python
AT07[['MontoInicial', 'MontoActual', 'MontoAvaluo', 'ValorMercado']] = AT07[['MontoInicial', 'MontoActual', 'MontoAvaluo', 'ValorMercado']].replace(to_replace=',', value='.', regex=True)
```

```python
AT07.MontoInicial = pd.to_numeric(AT07.MontoInicial, errors='coerce', downcast='float')
AT07.MontoActual = pd.to_numeric(AT07.MontoActual, errors='coerce', downcast='float')
AT07.MontoAvaluo = pd.to_numeric(AT07.MontoAvaluo, errors='coerce', downcast='float')
AT07.ValorMercado = pd.to_numeric(AT07.ValorMercado, errors='coerce', downcast='float')
```
