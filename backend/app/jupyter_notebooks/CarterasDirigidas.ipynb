```python
import pandas as pd
```

```python
pd.options.display.max_columns = 500
pd.options.display.max_rows = 10000
```

```python
parse_dates = [
    'FECHA_SOLICITUD',
    'FECHA_APROBACION',
    'FECHA_LIQUIDACION',
    'FECHA_VENC_ORIGINAL',
    'FECHA_VENC_ACTUAL',
    'FECHA_REESTRUCTURACION',
    'FECHA_PRORROGA',
    'FECHA_ULTIMA_RENOVACION',
    'FECHA_CANCEL',
    'FECHA_VENC_ULTIMA_CUOTA_CAPITAL',
    'ULTIMA_FECHA_CANCEL_CUOTA_CAPITAL',
    'FECHA_VENC_ULTIMA_CUOTA_INTERES',
    'ULTIMA_FECHA_CANCEL_CUOTA_INTERES',
    'FECHA_ESTADO_FINANCIERO',
    'FECHA_EMISION_FACTIBILIDAD_TECNICA',
    'FECHA_AUTENTICACION',
    'FECHA_ULTIMA_INSPECCION',
    'FECHA_VENC_REGISTRO',
    'Upd_Date'
]
```

```python
HLP = 6 # Hipotecario Largo Plazo
HCP = 7 # Hipotecario Corto Plazo
TURISMO = 8 # Turismo
MICROFINANCIERO = 9 # Microfinanciero
MANUFACTURA = 10 # Manufactura
AGRICOLA_ICG = 11 # Agricola ICG
AGRICOLA_OTHER_ICG = 12 # Agricola Other ICG
```

```python
cd_agricola_gcg = pd.read_csv('../vol/web/media/insumos/automatic/RPT_STG_Dirigidas_AGRICOLA_CONSUMER.txt', sep='	', low_memory=False, parse_dates=parse_dates, )
cd_agricola_icg = pd.read_csv('../vol/web/media/insumos/automatic/RPT_STG_Dirigidas_AGRICOLA_CORPORATE.txt', sep='	', low_memory=False, parse_dates=parse_dates)
cd_hlp = pd.read_csv('../vol/web/media/insumos/automatic/RPT_STG_Dirigidas_HIPOTECARIO_LARGO_PLAZO.txt', sep='	', low_memory=False, parse_dates=parse_dates)
cd_hcp = pd.read_csv('../vol/web/media/insumos/automatic/RPT_STG_Dirigidas_HIPOTECARIO_CORTO_PLAZO.txt', sep='	', low_memory=False, parse_dates=parse_dates)
cd_manufactura = pd.read_csv('../vol/web/media/insumos/automatic/RPT_STG_Dirigidas_MANUFACTURA.txt', sep='	', low_memory=False, parse_dates=parse_dates)
cd_micro = pd.read_csv('../vol/web/media/insumos/automatic/RPT_STG_Dirigidas_MICROFINANCIERO.txt', sep='	', low_memory=False, parse_dates=parse_dates)
cd_turismo = pd.read_csv('../vol/web/media/insumos/automatic/RPT_STG_Dirigidas_TURISMO.txt', sep='	', low_memory=False, parse_dates=parse_dates)
```

```python
cd_agricola_gcg.drop(cd_agricola_gcg.columns[len(cd_agricola_gcg.columns)-1], axis=1, inplace=True)
cd_agricola_icg.drop(cd_agricola_icg.columns[len(cd_agricola_icg.columns)-1], axis=1, inplace=True)
cd_hlp.drop(cd_hlp.columns[len(cd_hlp.columns)-1], axis=1, inplace=True)
cd_hcp.drop(cd_hcp.columns[len(cd_hcp.columns)-1], axis=1, inplace=True)
cd_manufactura.drop(cd_manufactura.columns[len(cd_manufactura.columns)-1], axis=1, inplace=True)
cd_micro.drop(cd_micro.columns[len(cd_micro.columns)-1], axis=1, inplace=True)
cd_turismo.drop(cd_turismo.columns[len(cd_turismo.columns)-1], axis=1, inplace=True)
```

```python
cd_agricola_gcg['TYPE_CD'] = AGRICOLA_OTHER_ICG
cd_agricola_icg['TYPE_CD'] = AGRICOLA_ICG
cd_hlp['TYPE_CD'] = HLP 
cd_hcp['TYPE_CD'] = HCP
cd_manufactura['TYPE_CD'] = MANUFACTURA 
cd_micro['TYPE_CD'] = MICROFINANCIERO
cd_turismo['TYPE_CD'] = TURISMO
```

```python
cd = pd.concat([cd_agricola_gcg, cd_agricola_icg, cd_hcp, cd_hlp, cd_manufactura, cd_micro, cd_turismo], ignore_index=True)
```
