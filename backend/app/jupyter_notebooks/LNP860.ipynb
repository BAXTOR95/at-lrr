```python
names = ['P8FVUC', 'P8FCCC', 'P8FVUI', 'P8FCCI', 'P8FCTC']
lnp860.P8FVUC =  pd.to_datetime(lnp860.P8FVUC, format='%d%m%Y', errors='coerce')
lnp860.P8FCCC =  pd.to_datetime(lnp860.P8FCCC, format='%d%m%Y', errors='coerce')
lnp860.P8FVUI =  pd.to_datetime(lnp860.P8FVUI, format='%d%m%Y', errors='coerce')
lnp860.P8FCCI =  pd.to_datetime(lnp860.P8FCCI, format='%d%m%Y', errors='coerce')
lnp860.P8FCTC =  pd.to_datetime(lnp860.P8FCTC, format='%d%m%Y', errors='coerce')
lnp860[names] = lnp860[names].fillna(pd.to_datetime('1900-01-01'))
```

```python
lnp860.P8NRCV = pd.to_numeric(lnp860.P8NRCV, errors='coerce')
lnp860.P8NRCV.fillna(0, inplace=True)
lnp860.P8NRCV = lnp860.P8NRCV.astype("int64")
lnp860.P8MOIN = pd.to_numeric(lnp860.P8NRCV, errors='coerce')
lnp860.P8MOIN.fillna(0, inplace=True)
lnp860.P8MOIN = lnp860.P8MOIN.astype("int64")
```

```python
fields = ['P8MV30', 'P8MV60', 'P8MV90', 'P8MV12', 'P8MV18', 'P8MV1A', 'P8MVM1', 'P8RPCV', 'P8LINT', 'P8MOCA', 'P8MOIN']
lnp860[fields] = lnp860[fields].apply(lambda x:x/100)
lnp860.P8TINC = lnp860.P8TINC.apply(lambda x:x/1000000)
```
