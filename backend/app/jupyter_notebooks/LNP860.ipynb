```python
lnp860.P8FVUC =  pd.to_datetime(lnp860.P8FVUC, format='%d%m%Y', errors='coerce')
lnp860.P8FCCC =  pd.to_datetime(lnp860.P8FCCC, format='%d%m%Y', errors='coerce')
lnp860.P8FVUI =  pd.to_datetime(lnp860.P8FVUI, format='%d%m%Y', errors='coerce')
lnp860.P8FCCI =  pd.to_datetime(lnp860.P8FCCI, format='%d%m%Y', errors='coerce')
lnp860.P8FCTC =  pd.to_datetime(lnp860.P8FCTC, format='%d%m%Y', errors='coerce')
lnp860.P8FVUC.fillna(pd.to_datetime('1900-01-01'), inplace=True)
lnp860.P8FCCC.fillna(pd.to_datetime('1900-01-01'), inplace=True)
lnp860.P8FVUI.fillna(pd.to_datetime('1900-01-01'), inplace=True)
lnp860.P8FCCI.fillna(pd.to_datetime('1900-01-01'), inplace=True)
lnp860.P8FCTC.fillna(pd.to_datetime('1900-01-01'), inplace=True)
```

```python
lnp860.P8NRCV = pd.to_numeric(lnp860.P8NRCV, errors='coerce')
lnp860.P8NRCV.fillna(0, inplace=True)
lnp860.P8NRCV = lnp860.P8NRCV.astype("int64")
```

```python
lnp860.P8MV30 = lnp860.P8MV30.apply(lambda x:x/100)
lnp860.P8MV60 = lnp860.P8MV60.apply(lambda x:x/100)
lnp860.P8MV90 = lnp860.P8MV90.apply(lambda x:x/100)
lnp860.P8MV12 = lnp860.P8MV12.apply(lambda x:x/100)
lnp860.P8MV18 = lnp860.P8MV18.apply(lambda x:x/100)
lnp860.P8MV1A = lnp860.P8MV1A.apply(lambda x:x/100)
lnp860.P8MVM1 = lnp860.P8MVM1.apply(lambda x:x/100)
lnp860.P8RPCV = lnp860.P8RPCV.apply(lambda x:x/100)
lnp860.P8LINT = lnp860.P8LINT.apply(lambda x:x/100)
lnp860.P8MOCA = lnp860.P8MOCA.apply(lambda x:x/100)
lnp860.P8MOIN = lnp860.P8MOIN.apply(lambda x:x/100)
lnp860.P8TINC = lnp860.P8TINC.apply(lambda x:x/1000000)
```

```python
lnp860.to_csv('../vol/web/media/insumos/automatic/lnp860.txt', sep='~', date_format='%d/%m/%Y', index=False)
```
