```python
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_columns = 500
pd.options.display.max_rows = 10000
```

```python
converters = {
    'BranchId':str, 
}

parse_dates = [
    'OpenDate',
    'RecordDate',
    'MaturityDate',
    'CloseDate',
    'BlockCode1Date',
    'OrigOpenDate',
    'Fecha_Cambio_Status',
    'Fecha_Reg_Venc_Lit_cast',
    'Fecha_Exigibilidad_pago_ult_cuota',
    'Fecha_Fin_Periodo_gracia_Pago_interes',
    'Fecha_cambio_Capital_Transferido'
]
```

```python
siif = pd.read_csv('../vol/web/media/insumos/automatic/Reporte SIIF Final.txt', 
                   sep='	', 
                   low_memory=False, 
                   encoding = "latin",
                   parse_dates=parse_dates,
                   converters=converters,)
```

```python
siif
```

```python
siif.CreditLimit.replace(to_replace='Bs.S', value='', inplace=True, regex=True)
siif.Amt30DPD.replace(to_replace='Bs.S', value='', inplace=True, regex=True)
siif.Amt60DPD.replace(to_replace='Bs.S', value='', inplace=True, regex=True)
siif.Amt90DPD.replace(to_replace='Bs.S', value='', inplace=True, regex=True)
siif.Amt120DPD.replace(to_replace='Bs.S', value='', inplace=True, regex=True)
siif.Amt150DPD.replace(to_replace='Bs.S', value='', inplace=True, regex=True)
siif.Amt180DPD.replace(to_replace='Bs.S', value='', inplace=True, regex=True)
siif.Amt210DPD.replace(to_replace='Bs.S', value='', inplace=True, regex=True)
siif.SaldoCastigado.replace(to_replace='Bs.S', value='', inplace=True, regex=True)
siif.PrincipalBalance.replace(to_replace='Bs.S', value='', inplace=True, regex=True)

siif.CreditLimit = pd.to_numeric(siif.CreditLimit, errors='coerce')
siif.Amt30DPD = pd.to_numeric(siif.Amt30DPD, errors='coerce')
siif.Amt60DPD = pd.to_numeric(siif.Amt60DPD, errors='coerce')
siif.Amt90DPD = pd.to_numeric(siif.Amt90DPD, errors='coerce')
siif.Amt120DPD = pd.to_numeric(siif.Amt120DPD, errors='coerce')
siif.Amt150DPD = pd.to_numeric(siif.Amt150DPD, errors='coerce')
siif.Amt180DPD = pd.to_numeric(siif.Amt180DPD, errors='coerce')
siif.Amt210DPD = pd.to_numeric(siif.Amt210DPD, errors='coerce')
siif.SaldoCastigado = pd.to_numeric(siif.SaldoCastigado, errors='coerce')
siif.PrincipalBalance = pd.to_numeric(siif.PrincipalBalance, errors='coerce')
```
