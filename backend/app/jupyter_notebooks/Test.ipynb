```python
from os import listdir
from os.path import isfile, join

import pandas as pd
```

```python
sb_config_path = '../vol/sb_config/'
sb_files = [f for f in listdir(sb_config_path) if isfile(join(sb_config_path, f))]
```

```python
sb_files = pd.DataFrame(sb_files, columns=['File'])
```

```python
sb_files[['Name', 'Extention']] = sb_files.File.str.split('.', n=1, expand=True)
```
